```python
def add_book(catalog,book_id,title,author,year):
    catalog[book_id]=(title,author,year)

def borrow_book(catalog,borrowed_books,book_id):
    if book_id in catalog and book_id not in borrowed_books:
        borrowed_books.append(book_id)
        print(f"Book {book_id} borrowed")
    else:
        print("Book unavailable")

def return_book(borrowed_books,book_id):
    if book_id in borrowed_books:
        borrowed_books.remove(book_id)
        print(f"\nBook {book_id} returned")

def register_member(members,member_id):
    members.add(member_id)

def show_available(catalog,borrowed_books):
    print("\nAvailable Books:")
    for book_id, details in catalog.items():
        if book_id not in borrowed_books:
            print(book_id, details)

def main():
    catalog = {}
    borrowed_books = []
    members = set()

    add_book(catalog,1,"Artificial Intelligence", "Stuart Russell", 2021)
    add_book(catalog,2,"Computer Networks","Andrew S. Tanenbaum",2018)
    add_book(catalog,3,"Database System Concepts", "Henry F. Korth", 2019)
    add_book(catalog,4,"Data Structures and Algorithms", "Narasimha Karumanchi", 2022)
    add_book(catalog,5,"Deep Learning","Ian Goodfellow",2016)
    add_book(catalog,6,"Java: The Complete Reference", "Herbert Schildt", 2021)
    add_book(catalog,7,"Machine Learning", "Tom M. Mitchell", 2017)
    add_book(catalog,8,"Operating System Concepts", "Abraham Silberschatz", 2020)
    add_book(catalog,9,"Python Programming", "John Zelle", 2020)


    register_member(members, 101)
    register_member(members, 102)
    register_member(members, 103)
    register_member(members, 104)
    register_member(members, 105)

    borrow_book(catalog, borrowed_books, 1)
    borrow_book(catalog, borrowed_books, 3)
    borrow_book(catalog, borrowed_books, 5)

    return_book(borrowed_books,3)

    show_available(catalog,borrowed_books)

    print("\nMembers:",members)
    print("\nBorrowed Books:",borrowed_books)

main()
```
