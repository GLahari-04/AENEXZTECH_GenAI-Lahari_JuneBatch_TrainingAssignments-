```python
from google import genai
from google.genai import types
from google.colab import userdata
```

```python
API_KEY = userdata.get("GEMINI_API_KEYS")

client = genai.Client(api_key=API_KEY)

print("API Connected Successfully!")
```

```python
knowledge_base = """
1. What courses are offered?
The CSE department offers B.Tech in Computer Science and Engineering.

2. What programming languages are taught?
Students learn C, C++, Java, Python, SQL, HTML, CSS, and JavaScript.

3. What are the laboratory facilities?
Programming Lab, AI & ML Lab, IoT Lab, Cloud Computing Lab, and Project Lab.

4. What are the department working hours?
Monday to Friday, 9:00 AM to 5:00 PM.

5. Who can I contact for admission enquiries?
Please contact the Admission Office or visit the CSE department office.

6. Does the department provide internships?
Yes. Internship opportunities are provided through the placement cell.

7. What companies visit for placements?
TCS, Infosys, Wipro, Cognizant, Accenture, Capgemini, and HCL.

8. What is the minimum attendance required?
Students must maintain at least 75% attendance.

9. Are project facilities available?
Yes. Students can work on AI, Web Development, Cloud Computing, Cyber Security, and Data Science projects.

10. Does the department conduct workshops?
Yes. Technical workshops, seminars, hackathons, and coding competitions are conducted regularly.

11. How can I access study materials?
Study materials are available on the department LMS portal.

12. What is the department email?
cse@college.edu

13. Where is the department located?
Block B, Second Floor.

14. Does the department provide Wi-Fi?
Yes. Free Wi-Fi is available throughout the department.

15. What placement training is provided?
Students receive aptitude, coding, communication, and interview preparation training.
"""
```

```python
system_instruction = f"""
You are the official FAQ chatbot for the Computer Science and Engineering Department.

Knowledge Base:
{knowledge_base}

Rules:

1. Answer ONLY using the knowledge base.

2. If the answer is not available in the knowledge base, reply:

"Sorry, I can only answer questions related to the CSE Department FAQ."

3. Be polite.

4. Keep answers short and clear.
"""
```

```python
chat = client.chats.create(
    model="gemini-2.5-flash",
    config=types.GenerateContentConfig(
        system_instruction=system_instruction,
        temperature=0.2
    )
)

print("Chatbot is Ready!")
```

```python
print("====================================")
print(" CSE Department FAQ Chatbot ")
print("Type 'quit' or 'exit' to stop.")
print("====================================")

while True:

    question = input("\nYou: ")

    if question.lower() in ["quit", "exit"]:
        print("\nBot: Thank you! Have a great day.")
        break

    response = chat.send_message(question)

    print("\nBot:", response.text)
```
