```python
from google.colab import userdata
from google import genai
from google.genai import types

# Load API Key from Colab Secrets
API_KEY = userdata.get("GEMINI_API_KEYS")

client = genai.Client(api_key=API_KEY)
```

```python
def generate_blog(topic, tone, word_count):

    system_prompt = """
    You are an experienced blog writer.

    Write a well-structured blog.

    The blog must include:
    1. Attractive Title
    2. Introduction Hook
    3. 2-3 Body Sections
    4. Conclusion

    Keep the language engaging and easy to understand.
    """

    user_prompt = f"""
    Topic: {topic}

    Tone: {tone}

    Word Count: {word_count}

    Generate the complete blog.
    """

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=user_prompt,
        config=types.GenerateContentConfig(
            system_instruction=system_prompt,
            temperature=0.7
        )
    )

    blog = response.text

    # Save to text file
    with open("blog_output.txt", "w", encoding="utf-8") as file:
        file.write(blog)

    print("\nBlog generated successfully!")
    print("Saved as blog_output.txt\n")

    return blog
```

```python
def main():

    print("========== BLOG GENERATOR ==========\n")

    topic = input("Enter Blog Topic: ")
    tone = input("Enter Tone (formal/casual/Gen Z/professional): ")
    word_count = input("Enter Word Count: ")

    blog = generate_blog(topic, tone, word_count)

    print("\nGenerated Blog\n")
    print("-" * 60)
    print(blog)
```

```python
main()
```
