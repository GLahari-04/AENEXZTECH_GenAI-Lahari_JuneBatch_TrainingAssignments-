```python
# Part A-Bug Demonstration
def add_item_bug(item,cart=[]):
    cart.append(item)
    return cart

print("Bug version")
print(add_item_bug("apple"))
print(add_item_bug("banana"))
print(add_item_bug("milk",cart=["bread"]))
print(add_item_bug("eggs"))


def add_item_fixed(item, cart=None):
    if cart is None:
        cart = []

    cart.append(item)
    return cart

print("\nFixed Version")
print(add_item_fixed("apple"))
print(add_item_fixed("banana"))
print(add_item_fixed("milk", cart=["bread"]))
print(add_item_fixed("eggs"))

# Part B - Fixed Version
def add_item(item,cart=None):
    if cart is None:
        cart=[]

    cart.append(item)
    return cart

# Part C
def create_cart(owner,discount=0):
    return {
        "owner":owner,
        "items":[],
        "discount":discount
    }

def add_to_cart(cart,name,price,qty=1):
    cart["items"].append({
        "name":name,
        "price":price,
        "qty":qty
    })

def update_price(price_tuple,new_price):
    try:
        price_tuple[0]=new_price
    except TypeError:
        print("TypeError: Tuple is immutable")
# Tuples are immutable,so their elements cannot be changed after creation.

def calculate_total(cart):
    total = 0

    for item in cart["items"]:
        total+=item["price"]*item["qty"]

    discount_amount=total*(cart["discount"]/100)

    return total-discount_amount

cart1=create_cart("Lahari",15)
cart2=create_cart("Manasa",5)

add_to_cart(cart1,"Books",50,15)
add_to_cart(cart1,"Pens",10,50)

add_to_cart(cart2,"Pencils",90,5)

print("\nCart 1")
print(cart1)
print("Total:",calculate_total(cart1))

print("\nCart 2")
print(cart2)
print("Total:",calculate_total(cart2))

price_info = (10,)
update_price(price_info, 20)

# Discussion Points:
# 1. discount=0 is safe because numbers cannot be changed(integers are immutable) and cart=[] is dangerous because lists(mutable) can be changed and shared.
# 2. Rebinding means giving a variable a new value.Mutating means changing the existing value.
# 3. Mutable:list,dict,set;Immutable:tuple,str,int
# 4. Yes,changes are reflected outside the function.This happens because the same list is used inside and outside the function.
```
